```python
# Databricks notebook source

 

# MAGIC %run /Users/saisudheerdara@gmail.com/common/03_Logger

import logging
from pyspark.sql.functions import *

logger = logging.getLogger("SocialMediaPipeline")
logger.setLevel(logging.INFO)

try:

    logger.info("Gold Tables ADLS Export Started")
    print("Gold Tables ADLS Export Started")

    gold_tables = [
        "stg_sentiment",
        "stg_trends",
        "stg_tweets",
        "stg_user_metadata",
        "stg_valid_tweets",
        "fact_social_media",
        "dim_date",
        "dim_sentiment",
        "dim_topic",
        "dim_user",
        "kpi_dashboard",
        "mart_country_summary",
        "mart_daily_summary",
        "mart_topic_summary"
    ]

    for table in gold_tables:

        logger.info(f"Reading Gold Table : {table}")
        print(f"Reading Gold Table : {table}")

        df = spark.table(f"socialmedia_databricks.gold.{table}")

        logger.info(f"Writing {table} to ADLS")
        print(f"Writing {table} to ADLS")

        (
            df.write
              .format("delta")
              .mode("overwrite")
             .save(f"abfss://raw@socialmedia1.dfs.core.windows.net/gold/{table}")
        )

        logger.info(f"{table} exported successfully")
        print(f"{table} exported successfully")

    logger.info("All Gold Tables Exported Successfully")
    print("All Gold Tables Exported Successfully")

except Exception as e:

    logger.error(f"Gold Tables Export Failed : {str(e)}")
    print(f"Gold Tables Export Failed : {str(e)}")

    raise
```
